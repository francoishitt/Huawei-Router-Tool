```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Define the tasks and their start/end months
tasks = [
    "Revisión bibliográfica y planificación",
    "Recolección de datos (imágenes, drones, campo)",
    "Procesamiento de datos y análisis",
    "Validación de resultados",
    "Redacción del informe",
    "Corrección y difusión"
]

# Ajuste de los meses para un cronograma de 18 meses
start_months = [1, 4, 7, 10, 14, 16]
end_months = [3, 6, 9, 13, 15, 18]  # Duraciones ajustadas para las actividades

# Define colors for the tasks
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", "#8c564b"]

# Create the DataFrame for Gantt chart
df = pd.DataFrame({
    "Task": tasks,
    "Start": start_months,
    "End": end_months,
    "Color": colors
})

# Plot the Gantt chart
fig, ax = plt.subplots(figsize=(12, 6))
for i, task in enumerate(df["Task"]):
    ax.barh(task, df["End"][i] - df["Start"][i] + 1, left=df["Start"][i], color=df["Color"][i])

# Format the chart
ax.set_xlabel("Meses")
ax.set_ylabel("Actividades")
ax.set_xticks(np.arange(1, 19, 1))
ax.set_xticklabels([f" {i}" for i in range(1, 19)])
ax.grid(axis='x', linestyle='--', alpha=0.7)
plt.title("Cronograma de Tesis (Diseño Gantt - 18 Meses)")
plt.tight_layout()

# Show the chart
plt.show()
```
